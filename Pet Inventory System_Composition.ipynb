```python
#Pet Inventory System
#Using Pet Composition

class Person:
    def __init__(self, owner_name: str, owner_address: str):
        self.name = owner_name
        self.address = owner_address
    
    def get_name_p(self):
        return self.name
    def get_address(self):
        return self.address
    
    def set_name_p(self,owner_name):
        self.name = owner_name
    def set_address(self,owner_address):
        self.address = owner_address
    
    def __str__(self):
        return 'Person name: ' + self.name + ", Address: " + self.address
```

```python
class PetComposition:
    def __init__(self, name: str, age: int, birth_weight: int, owner_name, owner_address):
        self.name = name
        self.age = age
        self.birth_weight = birth_weight
        self.owner = Person (owner_name, owner_address)
        #Here we create the object inside the __init__ method
    #Accessor Methods
    def get_name(self):
        return self.name
    def get_age(self):
        return self.age
    def get_birth_weight(self):
        return self.birth_weight
    def get_owner(self):
        return self.owner
    
    #Mutator Methods
    def set_name(self,name):
        self.name = name
    def set_age(self,age):
        self.age = age
    def set_birth_weight(self,birth_weight):
        self.birth_weight=birth_weight
    def __str__(self):
        pet = "pet's name: " + self.name + ", pet's age: " + str (self.age) + "pet's weight: " + str(self.birth_weight)
        pet += "pet's owner name: " + self.owner.get_name_p() + "pet's owner address:" + self.owner.get_address()
        return pet
```

```python
class Mammal(PetComposition):
    def __init__(self, name, age, birth_weight, owner_name, owner_address, litter_size:int, has_claws:bool):
        super().__init__(name, age, birth_weight, owner_name, owner_address)
        self.litter_size = litter_size
        self.has_claws = has_claws

    #Accessor Methods
    def get_litter_size(self):
        return self.litter_size
    def get_has_claws(self):
        return self.has_claws
    def get_weight(self):
        if self.age <=59:
            return self.birth_weight
        elif self.age <= 119:
            return self.birth_weight*(1.1**1)
        elif self.age <= 179:
            return self.birth_weight*(1.1**2)
        elif self.age <= 239:
            return self.birth_weight*(1.1**3)
        elif self.age <= 299:
            return self.birth_weight*(1.1**4)
        else:
            return self.birth_weight*(1.1**5)
    
    #Mutator Methods
    def set_litter_size(self, litter_size):
        self.litter_size = litter_size
    def set_has_claws(self, has_claws):
        self.has_claws = has_claws
        
    def __str__(self):
        pet = "Pet's name: " + self.name + ", pet's age: " + str (self.age) + ", pet's weight: " + str(self.get_weight())
        pet += ", pet's owner name: " + str(self.owner.get_name_p()) + ", pet's owner address: " + str(self.owner.get_address())
        pet += ", litter_size: " + str(self.litter_size) + ", claws available: " + str(self.has_claws)
        return pet
```

```python
class Fish(PetComposition): 
    def __init__(self, name, age, birth_weight, owner_name, owner_address, scale_condition:str, length:float):
        super().__init__(name, age, birth_weight, owner_name, owner_address)
        self.scale_condition = scale_condition
        self.length = length

   #Accessor Methods
    def get_scale_condition(self):
        return self.scale_condition
    def get_length(self):
        return self.get_length 
    def get_weight(self):
        if self.age <=79:
            return self.birth_weight
        elif self.age <= 159:
            return self.birth_weight*(1.05**1)
        elif self.age <= 239:
            return self.birth_weight*(1.05**2)
        else:
            return self.birth_weight*(1.05**3)

    #Mutator Methods
    def set_scale_condition(self, scale_condition):
        self.scale_condition = scale_condition
    def set_length(self,length):
        self.length = length      
        
    def __str__(self):
        pet = "Pet's name: " + self.name + ", pet's age: " + str (self.age) + ", pet's weight: " + str(self.get_weight())
        pet += ", pet's owner name: " + str(self.owner.get_name_p()) + ", pet's owner address:" + str(self.owner.get_address())
        pet += ", scale_condition: " + self.scale_condition + ", length: " + str(self.length)
        return pet
```

```python
class Amphibians(PetComposition): 
    def __init__(self, name, age, birth_weight, owner_name, owner_address, is_venomous: bool):
        super().__init__(name, age, birth_weight, owner_name, owner_address)
        self.is_venomous = is_venomous

    #Accessor Methods
    def get_is_venomous(self):
        return self.is_venomous
    def get_weight (self):
        if self.age <=119:
            return self.birth_weight
        elif self.age <= 239:
            return self.birth_weight*(1.05**1)
        elif self.age <= 359:
            return self.birth_weight*(1.05**2)
        else:
            return self.birth_weight*(1.05**3)

    #Mutator Methods
    def set_is_venomous(self, is_venomous):
        self.is_venomous = is_venomous
        
    def __str__(self):
        pet = "Pet's name: " + self.name + ", pet's age: " + str (self.age) + ", pet's weight: " + str(self.get_weight())
        pet += ", pet's owner name: " + self.owner.get_name_p() + ", pet's owner address:" + self.owner.get_address()
        pet += ", length: " + str(self.is_venomous)
        return pet
```

```python
import pickle
import sys

def main():
    
    while True:
        print('------------Menu------------')
        print('1. Add a new pet')
        print('2. Compute and print current weight of all pets in the file')
        print('3. Print all pet information for every pet in the file')
        print('4. Exit the application')

        choice = input("Your selection: ")
        try:
            choice = int(choice)
        except ValueError:
            print("You must enter an integer!")
            continue
        
        if choice not in [1,2,3,4]:
            print("Enter one of the above options only!")
            continue

        if choice == 1:
            type_of_pet = int(input("Type of pet?(1- Mammal; 2- Fish; 3- Amphibian) : "))
            
            try:
                type_of_pet = int(type_of_pet)
            except ValueError:
                print("Wrong Selection!")
            if type_of_pet not in [1,2,3]: 
                print("Wrong Selection!")
                main()
                
            pet_name = input("Enter the pet's name: ")
            pet_age = int(input("Enter the pet's age: "))
            pet_birthweight = input("Enter the pet's birth weight: ")
            
            try:
                pet_age = int(pet_age)
            except ValueError:
                print("Enter an Integer Value for Age!")
                print("Exiting the program")
                sys.exit()
            
            try:
                pet_birthweight = float(pet_birthweight)
            except ValueError:
                print("Enter a float value for Birthweight!")
                print("Exiting the program")
                sys.exit()
            
            owner_name = input("Enter Owner Name: ")
            owner_address = input("Enter Owner Address: ")

            if type_of_pet == 1:
                litter_size = int(input("Enter the litter size: "))
                has_claws = input("Does the pet have claws? Answer 'yes' or 'no' : ")
                
                if has_claws.upper() == 'YES':
                    has_claws = True
                elif has_claws.upper() == 'NO':
                    has_claws = False
                else:
                    print("Enter Yes or No! Please try again")
                    main()
                
                mammal = Mammal(pet_name, pet_age, pet_birthweight,owner_name, owner_address,litter_size, has_claws)

                with open('petdata.dat','ab') as f:
                    pickle.dump(mammal, f)              
                
            elif type_of_pet == 2:
                length = int(input("Enter the length of the fish in inches: "))
                scale = input("What is the condition of the scale? (Answer 'Rough' or 'Scaly':  ")
                
                if scale.upper() == 'ROUGH':
                    scale = "Rough"
                elif scale.upper() == 'SCALY':
                    scale = "Scaly"
                else:
                    print("Enter Rough or Scaly! Please try again")
                    main()
                
                fish = Fish(pet_name, pet_age, pet_birthweight,owner_name, owner_address, scale, length)

                with open('petdata.dat','ab') as f:
                    pickle.dump(fish, f)

            elif type_of_pet == 3:

                is_venomous = input("Is the pet venomous? Answer 'Yes' or 'No': ")
                
                if is_venomous.upper() == 'YES':
                    is_venomous = True
                elif has_claws.upper() == 'NO':
                    is_venomous = False
                else:
                    print("Enter Yes or No! Please try again")
                    main()
                
                amphibian = Amphibians(pet_name, pet_age, pet_birthweight,owner_name, owner_address,is_venomous)

                with open('petdata.dat','ab') as f:
                    pickle.dump(amphibian, f) 
                    
        elif choice==2:
            try:
                with open('petdata.dat', 'rb') as petdata:
                    try:
                        while(True):
                            pet = pickle.load(petdata)
                            petname = pet.get_name()
                            currentweight = pet.get_weight()
                            print(petname, 'weighs', currentweight, 'pounds.')
                    except EOFError:
                        pass
            except FileNotFoundError:
                print('No pets in inventory.') 
                continue
                  

        elif choice==3:
            try:
                with open('petdata.dat', 'rb') as fb:
                    try:
                        while (True):
                            auto = pickle.load(fb) 
                            print(auto)
                    except EOFError:
                        pass
            except FileNotFoundError:
                print('No pets in inventory.') 
                continue 

        elif choice==4:
            answer = input('You chose to exit the program\nAre you sure (Y/N)?')
            if answer == 'Y' or answer == 'y':
                sys.exit()
                
            elif answer == 'N' or answer == 'n':
                continue
    
main()
```
