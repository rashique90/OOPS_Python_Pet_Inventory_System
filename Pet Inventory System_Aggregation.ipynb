```python
#Pet Inventory System
#Using Pet Aggregation


import pickle
import sys


class Person:  # owner
    
    def __init__(self, owner_name, owner_address):
        self.__owner_name = owner_name
        self.__owner_address = owner_address

    #Accessor methods
    def get_owner_name(self):
        return self.__owner_name

    def get_address(self):
        return self.__owner_address
    
    #Mutator methods
    def set_owner_name(self, owner_name):
        self.__owner_name = owner_name

    def set_address(self, owner_address):
        self.__owner_address = owner_address
        
    #Other Methods    
    def current_weight(self):
        pass
        
    def get_name(self):
        pass

    def __str__(self): 
        return 'Owner name: ' + str(self.__owner_name) + "; Owner Address: " + str(self.__owner_address)


class PetAggregation:
    def __init__(self, name: str, age: int, birth_weight: float, owner):
        self.__name = name
        self.__age = age
        self.__birth_weight = birth_weight
        self.__person = owner

    #Accesor methods
    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def get_birth_weight(self):
        return self.__birth_weight

    def get_person(self):
        return self.__person
    
    #Mutator Methods
    def set_name(self, name):
        self.__name = name

    def set_age(self,age):
        self.__age = age

    def set_birth_weight(self, birth_weight):
        self.__birth_weight = birth_weight
        


    def __str__(self):
        return f'Pet name: ' + self.__name + ", Pet Age: " + str(self.__age) + ' Birth Weight: ' + str(self.__birth_weight) + '\nOwner Details \n' + self.__person.__str__()



class Mammal(PetAggregation):
 
    def __init__(self, name: str, age: int, birth_weight: float,litter_size: int, has_claws: bool, owner ):
        super(Mammal, self).__init__(name, age, birth_weight, owner)
        self.__litter_size = litter_size
        self.__has_claws = has_claws
    
    
    #Accesor methods
    def get_litter_size(self):
        return self.__litter_size

    def get_has_claws(self):
        return self.__has_claws
    
    #Mutator Methods
    def set_litter_size(self, litter_size):
        self.__litter_size = litter_size

    def set_has_claws(self, has_claws):
        self.__has_claws = has_claws
        
        
    def current_weight(self):
        if int(super().get_age()) <=59:
            return int(super().get_birth_weight())
        elif 60 <= int(super().get_age()) <= 119:
            return int(super().get_birth_weight())*(1.1**1)
        elif 120 <= int(super().get_age()) <= 179:
            return int(super().get_birth_weight())*(1.1**2)
        elif 180 <= int(super().get_age()) <= 239:
            return int(super().get_birth_weight())*(1.1**3)
        elif 240 <= int(super().get_age()) <= 299:
            return int(super().get_birth_weight())*(1.1**4)
        else:
            return int(super().get_birth_weight())*(1.1**5)
        

    def __str__(self):

        
        pet = "Details of this Mammmal pet are:\n" "Pet Name: " + super(Mammal,self).get_name() + ", Pet's age: " + str(super(Mammal,self).get_age()) +" days old;"+" Birth weight: " + str(super(Mammal,self).get_birth_weight())
        pet += "\nLitter_size: " + str(self.__litter_size) + "; Has Claws?: " + str(self.__has_claws)
        return pet


class Fish(PetAggregation):

    def __init__(self, name: str, age: int, birth_weight: float, length: float,scale_condition: str, owner ):
        super(Fish, self).__init__(name, age, birth_weight, owner)
        self.__scale_condition = scale_condition
        self.__length = length
    
        
    #Accesor methods
    def get_scale_condition(self):
        return self.__scale_condition

    def get_length(self):
        return self.__get_length
    
    #Mutator Methods
    def set_scale_condition(self, scale_condition):
        self.__scale_condition = scale_condition

    def set_length(self,length):
        self.__length = length
        
         
        
    def current_weight(self):
        if int(super().get_age()) <=79:
            return int(super().get_birth_weight())
        elif 80 <= int(super().get_age()) <= 159:
            return int(super().get_birth_weight())*(1.05**1)
        elif 160 <= int(super().get_age()) <= 239:
            return int(super().get_birth_weight())*(1.05**2)
        else:
            return int(super().get_birth_weight())*(1.05**3)
        
        
        
    

    def __str__(self):
        pet = "Details of this Fish pet are:\n" "Pet Name: " + super(Fish,self).get_name() + ", Pet's age: " + str (super(Fish,self).get_age()) +" days old;"+" Birth weight: " + str(super(Fish,self).get_birth_weight())
        pet += ", Scale_condition: " + str(self.__scale_condition) + ", Length: " + str(self.__length)
        return pet


class Amphibians(PetAggregation):
    
    
    def __init__(self, name: str, age: int, birth_weight: float,is_venomous: bool, owner ):
        super(Amphibians, self).__init__(name, age, birth_weight, owner)
        self.__is_venomous = is_venomous

    
    #Accesor methods
    def get_is_venomous(self):
        return self.__isVenomous
    
    #Mutator Methods
    def set_is_venomous(self, is_venomous):
        self.__is_venomous = is_venomous
        
    

    def current_weight(self):
        if int(super().get_age()) <=119:
            return int(super().get_birth_weight())
        elif 120 <= int(super().get_age()) <= 239:
            return int(super().get_birth_weight())*(1.05**1)
        elif 240 <= int(super().get_age()) <= 359:
            return int(super().get_birth_weight())*(1.05**2)
        else:
            return int(super().get_birth_weight())*(1.05**3)


    def __str__(self):
        pet = "Details of this Amphibian pet are:\n" "Pet Name: " + super(Amphibians,self).get_name() + ", Pet's age: " + str (super(Amphibians,self).get_age()) +" days old;"+" Birth weight: " + str(super(Amphibians,self).get_birth_weight())
        pet += ", Is Venomous: " + str(self.__is_venomous) 
        return pet




def main():
    #Start Here#
    '''Driver function'''
    
    '''
        Since the person(owner) object will be created inside the pet object, we just call the pet constructor
        with the instance variables required for the person(Owner) object
    '''
    #1 Using Aggregation, Owner object is first created before being passed as an argument to the Pet constructor
    
    
    while True:
        print("====Menu====")
        print("1. Add a new pet")
        print("2. Compute and print current weight of all pets in the file")
        print("3. Print all pet information for every pet in the file")
        print("4. Exit the application ")
        
        
        choice = input("Your selection: ")
        try:
           choice = int(choice)
        except ValueError:
            print("You must enter an integer!")
            continue
        
        if choice not in [1,2,3,4]:
            print("Enter one of the above options only!")
            continue
            
            

        
        if int(choice) == 1:
            type_of_pet = input("Type of pet?(1- Mammal; 2- Fish; 3- Amphibian) : ")
            
            #Validation for Pet Type
            try:
               type_of_pet = int(type_of_pet)
            except ValueError:
                print("Wrong Selection!")
            if type_of_pet not in [1,2,3]: 
                raise Exception("Wrong Selection!")
                
            
            pet_name = input("Enter the pet's name: ")
            pet_age = input("Enter the pet's age: ")
            pet_birthweight = input("Enter the pet's birth weight: ")
            
            #Validation for Age
            try:
               pet_age = int(pet_age)
            except ValueError:
                print("Enter an Integer Value for Age!")
                print("Exiting the program")
                sys.exit()
                
             #Validation for Birthweight
            try:
               pet_birthweight = float(pet_birthweight)
            except ValueError:
                print("Enter a float value for Birthweight!")
                print("Exiting the program")
                sys.exit()
                
                
            if type_of_pet not in [1,2,3]: 
                raise Exception("Wrong Selection!")
            
            
            owner_name = input("Enter Owner Name: ")
            owner_address = input("Enter Owner Address: ")
            
            owner = Person(owner_name, owner_address)
            
            
            if type_of_pet == 1:
                litter_size = int(input("Enter the litter size: "))
                has_claws = input("Does the pet have claws? Answer 'yes' or 'no' : ")
                
                #has_claws validation
                if has_claws.lower() == 'yes':
                    has_claws = True
                elif has_claws.lower() == 'no':
                    has_claws = False
                else:
                    print("Enter Yes or No")
                    print("Try Again!")
                    main()
                    
                    
                
                mammal = Mammal(pet_name, pet_age, pet_birthweight,litter_size, has_claws,owner)
                
                with open('petdata.dat','ab') as f:
                    pickle.dump(mammal, f)
                    pickle.dump(owner, f)
                

            elif type_of_pet == 2:
                length = int(input("Enter the length of the fish in inches: "))
                scale = input("What is the condition of the scale? (Answer 'Rough' or 'Scaly'):  ")
                
                #scale validation
                if scale.lower() == 'rough':
                    scale = "Rough"
                elif scale.lower() == 'scaly':
                    scale = "Scaly"
                else:
                    print("Enter Rough or Scaly")
                    print("Try Again!")
                    main()
                
                
                
                fish = Fish(pet_name, pet_age, pet_birthweight,length,scale,owner)
                with open('petdata.dat','ab') as f:
                    pickle.dump(fish, f)
                    pickle.dump(owner, f)

                
            elif type_of_pet == 3:

                is_venomous = input("Is the pet venomous? Answer 'Yes' or 'No': ")
                #is_venomous validation
                if is_venomous.lower() == 'yes':
                    is_venomous = True
                elif has_claws.lower() == 'no':
                    is_venomous = False
                else:
                    print("Enter Yes or No")
                    print("Try Again!")
                    main()
                
                amphibian = Amphibians(pet_name, pet_age, pet_birthweight,is_venomous,owner)

                with open('petdata.dat','ab') as f:
                    pickle.dump(amphibian, f)
                    pickle.dump(owner, f)
                
        elif choice==2:
            dict_pet = dict()
            try:
                with open('petdata.dat', 'rb') as petdata:
                    try:
                        print('Pet name and current weight of all Pets\n=========================================')
                        while(True):
                            pet = pickle.load(petdata)
                            petname = pet.get_name()
                            currentweight = pet.current_weight()
                            if petname == None or currentweight == None:
                                continue
                            else:
                                dict_pet[str(petname)] = str(currentweight)
                            print(petname, 'weighs', currentweight, 'pounds.')
                    except EOFError: 
                        pass          
            except FileNotFoundError: 
                print('No pets in inventory. Start with option 1 in the Main Menu.') 
                continue
                
        elif choice==3:
            try:
                with open('petdata.dat', 'rb') as fb:
                    try: 
                        while (True): 
                            auto = pickle.load(fb) 
                            print(auto)


                    except EOFError: 
                        pass 
            except FileNotFoundError: 
                print('No pets in inventory. Start with option 1 in the Main Menu.') 
                continue

        elif choice==4:
            exit_choice = input('Are you sure you want to exit? Y or N: ')
            if exit_choice == 'Y' or exit_choice == 'y':
                sys.exit()
            
            elif exit_choice == 'N' or exit_choice == 'n':
                continue
                
            else:
                print('Please enter either Y or N.')
    
 
main()
    
```
